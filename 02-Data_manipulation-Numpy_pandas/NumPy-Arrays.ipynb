# Random

#### Numpy has lot of ways to create random number arrays

## Rand

#### Creates an array of the given shape and populates it with random samples from a uniform distributor [0,1)

```python
np.random.rand(2)#generating 
```

```python
np.random.rand(4)
```

```python
np.random.rand(5,5)
```

```python
#it can create 1 dimensional and 2 dimensional arrays
```

# Randint

#### Returns random integers where the lowest number is inclusive but the highest exclusive

```python
np.random.randint(1,100)
```

```python
np.random.randint(1,100,10)
```

## Seed

```python
np.random.seed(67)
np.random.rand(4)
```

```python
np.random.seed(11)
np.random.randint(1,100,4)
```
