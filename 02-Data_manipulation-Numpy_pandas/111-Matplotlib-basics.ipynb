```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('String title here')
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('String title here')
plt.xlim(0,6)
plt.ylim(0,12)
plt.show()
```

## Figsize  

```python
plt.figure(figsize=(12,8))
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('String title here')
plt.show()
```

## This is called multiple figures

```python
plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('2,3,1')

plt.subplot(2,3,2)
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('2,3,2')

plt.subplot(2,3,3)
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('2,3,3')

plt.subplot(2,3,4)
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('2,3,4')

plt.subplot(2,3,5)
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('2,3,5')

plt.subplot(2,3,6)
plt.plot(x,y)
plt.xlabel ('X axis label here')
plt.ylabel ("Y axis label here")
plt.title ('2,3,6')

#(rows, columns, spot#) the first spot is for rows, that number expresses how many rows there are. 
# The second spot expresses the amount of columns.
# The last spot is the spot #. It works by going left to right top to bottom


plt.tight_layout()
plt.show()
```

## Scatter plot

```python
x = np.random.randint(1, 101, 100)
y = np.random.randint(1, 101, 100)

plt.figure(figsize=(8,6))
plt.scatter(x,y)

plt.xlabel("Random X values")
plt.ylabel("Random Y values")
plt.title ("Random Scatter Plot (Using numpy)")

plt.show()
```

```python
df = pd.read_csv('tips.csv')
df
```

```python
x = df['total_bill']
y = df['tip']

plt.figure(figsize=(12,9))
plt.scatter(x,y)

plt.xlabel(" Total Bill ")
plt.ylabel(" Tips ")
plt.title ("Scatter Plot comparing Total Bill vs Tips(Using numpy)")

plt.show()
```

```python
x = df['size']
y = df['tip']

plt.figure(figsize=(12,9))
plt.scatter(x,y)

plt.xlabel(" Size ")
plt.ylabel(" Tips ")
plt.title ("Scatter Plot comparing Size of Party vs Tips(Using numpy)")

plt.show()
```

```python
x = df['price_per_person']
y = df['tip']

plt.figure(figsize=(12,9))
plt.scatter(x,y)

plt.xlabel(" Price Per Person ")
plt.ylabel(" Tips ")
plt.title ("Scatter Plot comparing Price Per Person vs Tips(Using numpy)")

plt.show()
```
